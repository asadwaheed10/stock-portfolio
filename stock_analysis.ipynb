```python
from pandas_datareader import data
import datetime
from bokeh.plotting import figure, show, output_file

start = datetime.datetime(2018, 3, 14)
end = datetime.datetime(2020, 7, 27)

df=data.DataReader(name = "MSFT", data_source = "yahoo", start=start, end=end)

date_inc = df.index[df.Open < df.Close]
date_dec = df.index[df.Open > df.Close]

def inc_dec(close, open):
    if close > open:
        value = "Increase"
    elif close < open:
        value = "Decrease"
    else:
        value = "Equal"
    return value

df["Status"] = [inc_dec(close, open) for close, open in zip(df.Close, df.Open)]
df["Middle"] = (df.Open + df.Close)/2
df["Height"] = abs(df.Open - df.Close)

p = figure(x_axis_type = 'datetime', width = 1000, height = 300, sizing_mode = "scale_width")
p.title.text = "Stock Portfolio" 
p.grid.grid_line_alpha = 0.3

hours_12 = 12*60*60*1000

p.segment(df.index, df.High, df.index, df.Low, color = "black")
p.rect(df.index[df.Status == "Increase"], df.Middle[df.Status == "Increase"], hours_12, df.Height[df.Status == "Increase"],
      fill_color = "#006400", line_color = "black")
p.rect(df.index[df.Status == "Decrease"], df.Middle[df.Status == "Decrease"], hours_12, df.Height[df.Status == "Decrease"],
      fill_color = "#DC143C", line_color = "black")

df2=data.DataReader(name = "BAC", data_source = "yahoo", start=start, end=end)

date_inc = df2.index[df2.Open < df2.Close]
date_dec = df2.index[df2.Open > df2.Close]

df2["Status"] = [inc_dec(close, open) for close, open in zip(df2.Close, df2.Open)]
df2["Middle"] = (df2.Open + df2.Close)/2
df2["Height"] = abs(df2.Open - df2.Close)

p.segment(df2.index, df2.High, df2.index, df2.Low, color = "black")
p.rect(df2.index[df2.Status == "Increase"], df2.Middle[df2.Status == "Increase"], hours_12, df2.Height[df2.Status == "Increase"],
      fill_color = "#006400", line_color = "black")
p.rect(df2.index[df2.Status == "Decrease"], df2.Middle[df2.Status == "Decrease"], hours_12, df2.Height[df2.Status == "Decrease"],
      fill_color = "#DC143C", line_color = "black")

df3=data.DataReader(name = "VFV.TO", data_source = "yahoo", start=start, end=end)

date_inc = df3.index[df3.Open < df3.Close]
date_dec = df3.index[df3.Open > df3.Close]

df3["Status"] = [inc_dec(close, open) for close, open in zip(df3.Close, df3.Open)]
df3["Middle"] = (df3.Open + df3.Close)/2
df3["Height"] = abs(df3.Open - df3.Close)

p.segment(df3.index, df3.High, df3.index, df3.Low, color = "black")
p.rect(df3.index[df3.Status == "Increase"], df3.Middle[df3.Status == "Increase"], hours_12, df3.Height[df3.Status == "Increase"],
      fill_color = "#006400", line_color = "black")
p.rect(df3.index[df3.Status == "Decrease"], df3.Middle[df3.Status == "Decrease"], hours_12, df3.Height[df3.Status == "Decrease"],
      fill_color = "#DC143C", line_color = "black")

output_file("CS.html")
show(p)
```
